```python
import pandas as pd
import requests
from io import BytesIO 
```

```python
url = 'https://getdaytrends.com/es/top/tweeted/day/'
headers = { 'Accept' : '*/*',
            'User-Agent' : 'Mozilla/5.0',
            'Refers' : 'http://www.nseindia.com',
            'Connection' : 'keep-alive'
          }

getContents = requests.get(url,headers=headers).content
dataFromHtm=pd.read_html(BytesIO(getContents))

```
